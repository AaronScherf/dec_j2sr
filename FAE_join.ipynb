```python
# Import Packages
import pandas as pd
import numpy as np
```

```python
# Import DEC_J2SR Data
dec_j2sr_data = pd.read_csv('Working_Data/dec_j2sr_data.csv', index_col = "Unique_ID")
dec_j2sr_data
```

```python
dec_j2sr_data['Contract_Grant_Number_split'] = dec_j2sr_data['Contract_Grant_Number'].astype(str).str.replace(' ~\|~_©_~\|~', ",").str.replace(" ", "")
grant_number_dummies = dec_j2sr_data['Contract_Grant_Number_split'].str.get_dummies(sep=',')
grant_number_dummies
```

```python
grant_number_dummies.apply(pd.Series.value_counts).iloc[1,].sort_values(ascending=False)
```

```python
grant_columns = dec_j2sr_data['Contract_Grant_Number_split'].str.split(",",expand=True)
grant_columns
```

```python
dec_j2sr_data = dec_j2sr_data.join(grant_columns)
```

```python
fae_disb_data = pd.read_csv('Raw_Data/FAE_Disbursements/USAID_disb_2016.csv')

for year in range(2017,2021):
    path = 'Raw_Data/FAE_Disbursements/USAID_disb_' + str(year) + '.csv'
    fae_disb_year = pd.read_csv(path)
    fae_disb_data = fae_disb_data.append(fae_disb_year,ignore_index=True)
    
fae_disb_data
```

```python
fae_disb_reduced = fae_disb_data.loc[:,['activity_id', 'activity_name', 'activity_project_number', 'activity_start_date', 'activity_end_date', 
                     'country_id', 'country_code', 'country_name', 'region_id', 'region_name', 'transaction_type_id', 
                     'transaction_type_name', 'fiscal_year', 'current_amount', 'constant_amount', 'activity_description']]
fae_disb_reduced.head()
```

```python
fae_disb_reduced['activity_project_country_id'] = fae_disb_reduced['activity_project_number'].astype(str) + "_" + fae_disb_reduced['country_id'].astype(str)
fae_disb_reduced.sort_values('activity_id_country_id')
```

```python
fae_disb_reduced.pivot(index='activity_id_country_id',columns='fiscal_year', values='current_amount')
```

```python
grant_columns.shape
```

```python
merged_data = grant_columns.merge(fae_disb_reduced, how='left', left_on=0, right_on="activity_project_number")
```

```python
merged_data.columns
```

```python
merged_data.dropna(subset=['activity_id','country_id',0],inplace=True)
merged_data['contract_country'] = merged_data[0].astype(str) + "_" + merged_data['country_id'].astype(str)
merged_data['contract_country']
```

```python
merged_data.pivot_table(index='contract_country', columns='fiscal_year', values='current_amount')
```

```python
merged_data.pivot_table(index='activity_project_number', columns='fiscal_year', values='current_amount')
```
